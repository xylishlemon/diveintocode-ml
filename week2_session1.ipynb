### 【問題1】1次関数
x
 が-50から50の範囲で次の式を考えます。  
 
$$y = \frac{1}{2}x+1$$  

-50から50まで0.1間隔のxを表すndarrayと、各要素に対するyを表すndarrayを作成してください。

```python
import numpy  as np
import matplotlib.pyplot as plt
# import japanize_matplotlib

a = 1/2 # 定数0.5を変数に格納
x = np.arange(-50, 50.1, 0.1) # -50から50までの配列
y = (a * x) + 1 # 1次関数y

print(x.shape)
print(y.shape)

print("配列xは、\n{}".format(x))
print("配列yは、\n{}".format(y))
```

### 【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。  0列目にx、1列目にyが来るようにします。

```python
line_x_y = np.hstack((x.reshape(1001,1), y.reshape(1001,1))) # x,yを(1001, 1)に変換

print("xとyの配列を水平方向に結合したもの\n{}".format(line_x_y))
print(line_x_y.shape)
# print(line_x_y.ndim)
```

### 【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。

![勾配](https://t.gyazo.com/teams/diveintocode/7fd04c4808b8a0150a49c908e21f3cd2.png) 
各変化量は以下の図のように隣同士の要素の差から近似的に求めます。

![勾配](https://t.gyazo.com/teams/diveintocode/398d8c5b4b1bd4151dea6488059226a1.png)
注意

- 厳密に各点における勾配を求めるには極限（微分）を利用する必要があります。  
- 勾配はx座標の間隔を使うので、xの要素数に対して一つ少なくなります。

```python
d = line_x_y[1:]- line_x_y[:-1] # 行列b[1:]とｂの[:-1]の差を出す
slope = d[:, 1] / d[:, 0] # xの変化量とyの変化量を出し、勾配を求める
#print(type(slope))
print("各x座標におけるxとyの変化量は、\n{}".format(slope))
```

### 【問題5】Pythonの関数化
問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。

そして、 $ y = x^2, y = 2x^2+2^x, y = sin(x^{\frac{1}{2}})$
 についての勾配を求めてください。
x
 の範囲は前2つが-50から50、最後の1つが0から50です。0.1間隔でndarrayを作成してください。

また、問題4と同様にグラフを描いてください。

ndarrayに対して三角関数のsinの計算を行うには np.sin()を用います。

[numpy.sin — NumPy v1.16 Manual]
(https://docs.scipy.org/doc/numpy/reference/generated/numpy.sin.html#numpy.sin)

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
  変化量を使い勾配を計算する。

  Parameters
  ----------------
  function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
  x_range : tuple
    np.arange()の引数と同じ方法で範囲を指定する。

  Returns
  ----------------
  array_xy : ndarray, shape(n, 2)
    xとyを結合したもの。nはx_rangeに依存する。
  gradient : ndarray, shape(n-1,)
    関数の勾配。差をとるため要素は1つ減る
    """
    a1, a2, a3 = x_range # tuple→intに変換
    x1 = np.arange(a1, a2, a3) # xの範囲
    c = x1.size # xの要素数（=yの要素数）
    y = function # 引数（function）はそれぞれの計算結果（=y）
    array_xy = np.hstack((x1.reshape(c, 1), y.reshape(c, 1))) # yを(1001, 1)変換し水平結合したもの
    e = array_xy[1:] - array_xy[:-1]
    gradient = e[:, 1] / e[:, 0]  # xの変化量とyの変化量を出し、勾配を求める
   
    return array_xy, gradient


def function1(array_x1): # y = x**2の関数定義
    array_y = (array_x1)**2 # yの配列
    return array_y # yの配列を返す

def function2(array_x1): # y = 2x**2 + 2**xの関数定義
    array_y = 2*array_x1**2 + 2**array_x1 # yの配列
    return array_y

def function3(array_x2): # y = sin(x**1/2)の関数定義
    array_y = np.sin(array_x2**(1/2)) # yの配列
    return array_y

array_x1 = np.arange(-50, 50.1, 0.1) # -50から50の配列
array_x2 = np.arange(0, 50.1, 0.1) # 0から50の配列

f_1 = function1(array_x1) # 定義した関数を変数に格納
f_2 = function2(array_x1) # 定義した関数を変数に格納
f_3 = function3(array_x2) # 定義した関数を変数に格納

np.set_printoptions(precision=3) # 小数点以下桁数の設定

arr = (0, 50.1, 0.1) # f_3のみ範囲が違いので設定

# print(compute_gradient(f_1))
# print(compute_gradient(f_2))
print(compute_gradient(f_3, x_range = arr)[1].shape)

#ray_xy, gradient = compute_gradient(function1)
```

```python
plt.figure(figsize=(8, 5), dpi=70)
plt.rcParams["font.size"] = 15
plt.title("$y=x^2$")
plt.plot(array_x1, function1(array_x1), label="$y=x^2$")
# compute_gradient(f_1)[1]は戻り値の2つ目を取得
plt.plot(array_x1[:-1], compute_gradient(f_1)[1], label="Gradient")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5), dpi=70)
plt.rcParams["font.size"] = 15
plt.title("$y=2x^2+2^x$")
plt.plot(array_x1, function2(array_x1), label="$y=2x^2+2^x$")
plt.plot(array_x1[:-1], compute_gradient(f_2)[1], label="Gradient")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5), dpi=70)
plt.rcParams["font.size"] = 15
plt.title("$sin(x^{1/2})$")
plt.plot(array_x2, function3(array_x2), label="$sin(x^{1/2})$")
plt.plot(array_x2[:-1], compute_gradient(f_3, arr)[1], label="Gradient")
plt.legend()
plt.show()
```

### 【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。

今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。

また、最小のyをとる前後の勾配を表示してください。

numpy.ndarray.min — NumPy v1.16 Manual

numpy.ndarray.argmin — NumPy v1.16 Manual

```python
np.set_printoptions(precision=5)
print("関数x^2の最小値とインデックスは: {},  {}".format(np.min(function1(array_x1)), np.argmin(function1(array_x1))))
print("関数2x^2+2^xの最小値は: {}  {}".format(np.min(function2(array_x1)), np.argmin(function2(array_x1))))
print("関数sin(x**1/2)の最小値の最小値は: {}  {}".format(np.min(function3(array_x2)), np.argmin(function3(array_x2))))
print("----------------")
np.set_printoptions(precision=3)
print("x^2の最小値のyをとる前後の勾配は、{}と{}".format(compute_gradient(f_1)[1][499], compute_gradient(f_1)[1][501]))
print("2x^2+2^xの最小値のyをとる前後の勾配は、{}と{}".format(compute_gradient(f_2)[1][497], compute_gradient(f_2)[1][499]))
print("sin(x**1/2)の最小値のyをとる前後の勾配は、{}と{}".format(compute_gradient(f_3,arr)[1][221], compute_gradient(f_3, arr)[1][223]))
```
