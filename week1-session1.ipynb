### 【問題1】100日後の米粒の数
100日後には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
#【問題1】100日後の米粒の数
import matplotlib.pyplot as plt
%matplotlib inline

today_list = []
sum_list = []
date = 100

rice_today = 1
rice_sum = 1
today_list = [rice_today]
sum_list = [rice_sum]

for i in range(date-1):
    rice_today *= 2
    today_list.append(rice_today)
    rice_sum += rice_today
    sum_list.append(rice_sum)
    
#print("その日にもらう米の数：{}".format(today_list))
#print("その日までに累計で貰う米の数：{}".format(sum_list))

#グラフ推移：その日にもらう米の数
plt.title("number of rice after 100 days")
plt.xlabel("date")
plt.ylabel("number of rice")
plt.plot(today_list)
plt.plot(sum_list)
plt.show()
```
