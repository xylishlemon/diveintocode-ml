### 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
import numpy as np
 
mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)
print(values[0])
print(values.shape)
```

### 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)

plt.scatter(values[:,0], values[:,1])
plt.show()
```

### 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)
values_max = np.max(values) # 最大値を格納
values_min = np.min(values) # 最小値を格納

# 列を指定せずにプロットした状態
plt.hist(values, bins=100)
plt.show()

# 1列目を指定してプロット
plt.hist(values[:,0], bins=100)
plt.xlim(values_min, values_max) # 最小値から最大値までで範囲を指定
plt.show()

# 2列目を指定してプロット
plt.hist(values[:,1], bins=100)
plt.xlim(values_min, values_max)
plt.show()
```

### 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。  
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
mu1 = [-3, 0]
sigma1 = [[1.0, 0.8], [0.8, 1.0]]

mu2 = [0, -3]
sigma2 = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values1 = np.random.multivariate_normal(mu1, sigma1, 500)
values2 = np.random.multivariate_normal(mu2, sigma2, 500)

plt.scatter(values1[:,0], values1[:,1], marker='o')
plt.scatter(values2[:,0], values2[:,1], marker='o')
plt.show()
```

### 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。  
結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
mu1 = [-3, 0]
sigma1 = [[1.0, 0.8], [0.8, 1.0]]

mu2 = [0, -3]
sigma2 = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values1 = np.random.multivariate_normal(mu1, sigma1, 500)
values2 = np.random.multivariate_normal(mu2, sigma2, 500)

# データ(NumPy配列)をnp.concatenateを用いて結合
connect_values = np.concatenate([values1, values2], axis=0)
print(connect_values)
```

### 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
mu1 = [-3, 0]
sigma1 = [[1.0, 0.8], [0.8, 1.0]]

mu2 = [0, -3]
sigma2 = [[1.0, 0.8], [0.8, 1.0]]

# 乱数のシード値を固定
np.random.seed(0)

# 2次元正規乱数を500個生成
values1 = np.random.multivariate_normal(mu1, sigma1, 500)
values2 = np.random.multivariate_normal(mu2, sigma2, 500)

zero = np.zeros(500) # ０配列の生成
one = np.ones(500) # 1配列の生成

connect_values1 = np.c_[values1, zero] # 0と結合
connect_values2 = np.c_[values2, one] # 1と結合

# データ(NumPy配列)をnp.concatenateを用いて結合
connect_values = np.concatenate([connect_values1, connect_values2], axis=0)

print(connect_values)
plt.scatter(connect_values[:,0], connect_values[:,1], s=50, c=connect_values[:,2], marker='o')
plt.show()
```
