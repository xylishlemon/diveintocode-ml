DataFrame のデータを参照する方法はいくつかあるが、その中にloc、iloc、ix というよく似た性質のものがあります。この3つは、行、列 を指定してデータを参照するという、基本的に同じ目的で使われます。  
異なる点は 行、列の指定の方法で、次のような違いがあります。

- loc：行ラベル、 列ラベルを指定
- iloc：行の番号(0 ~ )、列の番号(0 ~ )を指定
- ix：行ラベル、列ラベル または 行の番号(0 ~ )、列の番号(0 ~ )

###### ix の注意点
インデックスやカラムが整数型 だった場合、行の番号を指すのではなく、インデックスを指すことに注意が必要。

###### 箱ひげ図とバイオリン図の違いについて
- 箱ひげ図：データのばらつきをわかりやすく表現するための統計図。主に多くの水準からなる分布を視覚的に要約し、比較するために用いる。箱（box）と、その両側に出たひげ（whisker）で表現されることからこの名が付けられた。 

- バイオリン図：箱ひげ図に似ているが、異なる値のデータに確率密度を表示する点で異なっている。確率密度は、最も単純な場合にヒストグラムとなる。一般にバイオリン図には、箱ひげ図同様、データの中央値を示すマークと四分位範囲を示す箱も描かれる。この箱ひげ図に重ねて描画されるのはカーネル密度推定である。箱ひげ図のようにバイオリン図は複数のカテゴリ間での変数分布の比較に用いられる。

### 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。  

- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る  
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る  
- ４つの特徴量に関する相関係数行列を作る  
- 相関係数行列のヒートマップを作る  

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。

```python
# setosa、virgiclor、virginica（ラベル）
# sepal_length  sepal_width  petal_length  petal_width（特徴量）df.loc["2013-01-01"]
# >>> # 列 "A", "B" の 2 列を抽出
# >>> df.loc[:,['A','B']]　
#df[df['age'] < 25]
#df[df["petal_width"] == 0.2]

df_setosa = df[df["Species"] == 0]
df_virgiclor = df[df["Species"] == 1]
df_virginica = df[df["Species"] == 2]
# print(df_setosa.head())

#散布図
plt.figure(dpi=100)

plt.title("sepal_length  × sepal_width", fontsize=15)
plt.xlabel("sepal_length", fontsize=15)
plt.ylabel("sepal_width", fontsize=15)
plt.scatter(df_setosa.iloc[:, 0], df_setosa.iloc[:, 1], c='r', marker='.', s=15, label="setosa")
plt.scatter(df_virgiclor.iloc[:, 0], df_virgiclor.iloc[:, 1], c='b',marker='^', s=15, label="virgiclor")
plt.scatter(df_virginica.iloc[:, 0], df_virginica.iloc[:, 1], c='y',marker='s', s=15, label="virginica")
plt.legend()

plt.show()
```

```python
print("散布図行列を作る")
sns.pairplot(df,hue='Species', palette='magma')
```

```python
print("４つの特徴量に関する相関係数行列は")
df.corr(method='pearson')
```
